```python
# Základní importy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ISLP import load_data

# Načtení dat
advertising = pd.read_csv("Advertising.csv", index_col=0)
advertising = advertising.dropna()
# advertising.boxplot("sales", by="TV");

# Výpočet koeficientů
x = advertising["TV"]
y = advertising["sales"]

x_mean = x.mean()
y_mean = y.mean()

beta_1 = ((x - x_mean)*(y - y_mean)).sum()/((x - x_mean)**2).sum()
beta_0 = y_mean - beta_1*x_mean

print(f"beta_1 (strmost): {beta_1}")
print(f"beta_0 (protínání osy y): {beta_0}")

y_hat = beta_0 + beta_1 * x
residuals = y - y_hat

RSS = (residuals**2).sum()
print(f"RSS: {RSS}")

# Graf
plt.scatter(advertising["TV"], advertising["sales"], c="green")
plt.plot(x, y_hat, color="black", label="Regresní přímka")
plt.xlabel("TV reklama (tisíce $)")
plt.ylabel("Prodeje (tisíce jednotek)")
plt.title("Scatter plot TV reklama vs prodeje")
plt.show()
```
